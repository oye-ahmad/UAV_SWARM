```python
import cv2
```

```python
classes = {
    0: "Ignored",
    1: "Pedestrian",
    2: "People",
    3: "Bicycle",
    4: "Car",
    5: "Van",
    6: "Truck",
    7: "Tricycle",
    8: "Awning-tricycle",
    9: "Bus",
    10: "Motor Cycle",
    11: "Others"
}
```

```python
image_path = "C:\\Users\\MUHAMMADAHMAD\\Desktop\\InternShip Week 3\\0000002_00448_d_0000015.jpg"
annotation_path = "C:\\Users\\MUHAMMADAHMAD\\Desktop\\InternShip Week 3\\0000002_00448_d_0000015.txt"
```

```python
image = cv2.imread(image_path)
if image is None:
    print("Image not found!")
    exit()
```

```python
with open(annotation_path, "r") as file:
    annotations = file.readlines()
```

```python
for line in annotations:
    parts = line.strip().split(',')
    x = int(parts[0])
    y = int(parts[1])
    w = int(parts[2])
    h = int(parts[3])
    score = int(parts[4])
    category = int(parts[5])
    truncation = int(parts[6])
    occlusion = int(parts[7])
    if score == 0:
        continue  
    else:
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        label = classes.get(category, "Unknown")
        cv2.putText(image,label,(x, y - 5),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,0),2)
    
```

```python
cv2.imshow("VizDrone Annotation", image)

cv2.waitKey(0)

cv2.destroyAllWindows()
```
