```python
import cv2
import numpy as np 
```

```python
image = cv2.imread(r"C:\Users\MUHAMMADAHMAD\Desktop\Internship Week 1\TestImage.jpg")
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
annotated=image.copy()
cv2.rectangle(annotated, (380, 40), (560, 550), (0, 255, 0), 3)
cv2.putText(annotated, 'Tree Detected', (380, 30),
            cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)


cv2.circle(annotated, (300, 150), 40, (0, 200, 255), 3)
cv2.putText(annotated, 'Light Source', (250, 100),
            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 200, 255), 2)


cv2.line(annotated, (300, 150), (470, 300), (255, 0, 0), 2)


cv2.circle(annotated, (300, 150), 5, (0, 0, 255), -1)
cv2.imwrite('annotated_output.jpg', annotated)
```

```python
cv2.imshow("Gray Scale Version", gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blurred = cv2.GaussianBlur(image,(5,5),0)
```

```python
cv2.imshow("Gaussian Blur Version", blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edges = cv2.Canny(gray_image,100,200)
```

```python
cv2.imshow("Canny Version", edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
h, w, c = image.shape
crop = image[0:int(h/2), 0:int(w/2)]
# cropped = image[100:300, 150:400]

b, g, r = cv2.split(image)
merged = cv2.merge([b, g, r])
print(np.array_equal(merged, image)) 

pixel = image[50, 50]       
image[50, 50] = [255, 0, 0] 

resized = cv2.resize(image, (300, 300))

bright = cv2.add(image, np.ones(image.shape, dtype='uint8') * 50)
```
