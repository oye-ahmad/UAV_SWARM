```python
from ultralytics import YOLO

# Load your trained best weights
model = YOLO("/content/runs/detect/train/weights/best.pt")

# Run prediction on a test image
results = model.predict(
    source="/content/1.jpg",   # change this path
    save=True,
    conf=0.25,   # keep this lower to start; raise later if too many false boxes
    show_labels=True,
    show_conf=True
)

# Print how many chickens it found
for r in results:
    print(f"Detected {len(r.boxes)} chicken(s)")
    for box in r.boxes:
        print(f"Confidence: {box.conf.item():.2f}, Coordinates: {box.xyxy.tolist()}")
```
