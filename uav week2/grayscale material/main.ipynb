```python
import cv2
import numpy as np

# 1. Image load karein (Yahan apni image ka sahi path likhein)
image_path = r"C:\UAV\UAV_SWARM\uav week2\grayscale image1\image2.jpg"
frame = cv2.imread(image_path)

# Agar image load nahi hui to error show karein
if frame is None:
    print(f"Error: Image '{image_path}' nahi mili! Sahi path check karein.")
    exit()

# 1 se start hone wala frame/image counter (log ke liye)
frame_number = 1
frame_log = []

###################################################
# Grayscale Conversion (B&W)
###################################################
# Processing hamesha B&W image par behtar hoti hai
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

###################################################
# Gaussian Blur (Grayscale image par)
###################################################
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

###################################################
# Canny Edge Detection
###################################################
edges = cv2.Canny(blurred, 50, 100)

###################################################
# Contour Extraction
###################################################
contours, _ = cv2.findContours(
    edges,
    cv2.RETR_EXTERNAL,
    cv2.CHAIN_APPROX_SIMPLE
)

contour_image = frame.copy()

# Tamam contours ko green color (0, 255, 0) mein draw karein
cv2.drawContours(
    contour_image,
    contours,
    -1,
    (0, 255, 0),
    2
)

# Har contour ka area check karke detect kiye gaye objects ka count rakhna
detection_count = 0
for cnt in contours:
    area = cv2.contourArea(cnt)
    if area > 300:  # Sirf unhein count karein jo size mein bade hain
        detection_count += 1

###################################################
# Save Data
###################################################
frame_log.append({
    "Frame": frame_number,
    "Detection Count": detection_count
})

print(f"Image processed! Detections found: {detection_count}")

###################################################
# Show Windows
###################################################
cv2.imshow("Original", frame)
cv2.imshow("Grayscale", gray)
cv2.imshow("Blur", blurred)
cv2.imshow("Edges", edges)
cv2.imshow("Contours (Detected)", contour_image)

# Kisi bhi key ke press hone ka wait karein taake windows khuli rahein
cv2.waitKey(0)
cv2.destroyAllWindows()
```
